##Part 2

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, cos, sin, simplify, Eq, lambdify

phi = symbols('phi')

lhs = sin(phi)**2

rhs = 1/2 * (1 - cos(2*phi))

cos_2phi = 1 - 2*sin(phi)**2

rhs = rhs.subs(cos(2*phi), cos_2phi)

lhs_simplified = simplify(lhs)
rhs_simplified = simplify(rhs)

identity_proven = Eq(lhs_simplified, rhs_simplified)

phi_vals = np.linspace(0, 2*np.pi, 1000)
lhs_func = lambdify(phi, lhs, 'numpy')
rhs_func = lambdify(phi, rhs, 'numpy')
cos_2phi_func = lambdify(phi, cos_2phi, 'numpy')

plt.figure(figsize=(5, 5))
plt.title('LHS')
plt.plot(phi_vals, lhs_func(phi_vals), label='LHS: sin^2(phi)', linewidth=2)
plt.figure(figsize=(5, 5))
plt.title('RHS')
plt.plot(phi_vals, rhs_func(phi_vals), label='RHS: 1/2 * (1 - cos(2*phi))', linewidth=2)
plt.figure(figsize=(5, 5))
plt.plot(phi_vals, cos_2phi_func(phi_vals), label='Euler form of cos(2*phi)', linestyle='dashed', linewidth=2)

plt.title('Trigonometric Identity and Euler Form')
plt.legend()
plt.grid(True)
plt.show()

print(f"Left-hand side: {lhs_simplified}")
print(f"Right-hand side: {rhs_simplified}")
print(f"Identity is {identity_proven}")
```

```python
import numpy as np

def euler_identity(theta):
    return np.exp(1j * theta)

res = euler_identity(np.pi) + 1

if np.isclose(res, 0):
    print("Euler's Identity holds true: e^(i*pi) + 1 = 0")
else:
    print("Euler's Identity does not hold true.")
```
